# Write a funtion which takes n as input and returns a sum of first n natural numbers

```python
def sum_of_natural_numbers(n):

  if n <=0:
    return print("Invalid Input: Please enter a number greater than 0")

  sum = 0

  for i in range(1, n+1):
    sum += i

  return sum
```

```python
sum_of_natural_numbers(4)
```

```python
sum_of_natural_numbers(3)
```

```python
sum_of_natural_numbers(5)
```

```python
sum_of_natural_numbers(-4)
```

```python
sum_of_natural_numbers(500**5000)
```
